```python
import re
def registration():
    temp=0
    pattern = '^[a-z]+[\._]?[a-z 0-9]+[@]\w+[.]\w{2,3}$'
    
    username = input("Enter email id:")
    with open('newa.txt','r') as guvi:
        if re.search(pattern,username):
            for line in guvi:
                if username in line:
                    print("username available")
                    temp=1
                    break
        else:
            print("enter valid emailid")
            temp=1

    specials = ['@','!','#','^','%','$','&','*','(',')']
    if temp != 1:
        password=input('Enter password :')
        if len(password) <16 and len(password) >5:
            if re.search('[0-9]',password):
                if re.search('[A-Z]',password):          
                    if re.search('[a-z]',password):
                        if any(char in specials for char in password):
                            with open('newa.txt', 'a') as iit:
                                iit.write(f"\n{username},{password}")
                                print("Registration is Successful")
                        else:
                            print("Password should have special letter")

                    else:
                        print("Password should have lower case letter")
                else:
                    print("Password should have capital letter")
            else:
                print("Password should have Number")
        else:
            print('password length should be between 8 to 16')

        # forgot password function
def forgot_password():
    forgot_mail=input('Enter the registered mail id:\n')
    with open('newa.txt','r') as ch:
        for line in ch:
            if forgot_mail in line:
                pas=line.split(',')
                print("The Registered password is",pas[1])                              
                break
            else:
                print("Email Id is not valid")


   # login function
def login():
    login_mail=input('Login E-mail:')
    login_pass=input('Login Password:')
    
    with open('newa.txt','r') as tn:
        for line in tn:
            if login_mail and login_pass in line:
                escape = re.sub(r"(\\n|\\r|\\t|\\)", "", line).strip()
                check=escape.split(',')
                if login_mail ==check[0] and login_pass == check[-1]:
                    print("login Sucessful")
                    break
                else:
                    new_input=int(input("Incorrect Credentials ,select the following : \n 1.registration \n2.Forgot password"))
            
            if new_input == 1:
                registration()
            elif new_input == 2:
                forgot_password()
            else:
                print("kindly register to continue !")
                                
#Program_
user_input=int(input("Enter your action \n 1.Registration \n 2.Login \n 3.Forgot Password"))
if user_input == 1:
    registration()
elif user_input == 2:
    login()
elif user_input == 3:
    forgot_password()
else:
    print("Enter correct input !")
            
```
